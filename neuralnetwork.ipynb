Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs
from traditional machine learning.

Ans - Deep Learning is a subfield of machine learning that focuses on training artificial neural networks with many layers (deep neural networks) to automatically learn hierarchical representations from large amounts of data. It is particularly effective for complex tasks such as image recognition, speech processing, natural language understanding, and autonomous systems.

Evolution of Deep Learning -

Early Neural Networks (1940s–1980s)

Inspired by the human brain, early models like the Perceptron (Rosenblatt, 1958) attempted simple pattern recognition.

These models were limited to linear problems and lacked computational power and data.

Backpropagation Era (1980s–1990s)

The introduction of the backpropagation algorithm enabled multi-layer neural networks to be trained.

However, training deep networks was difficult due to problems like vanishing gradients, limited datasets, and insufficient hardware.

Revival of Deep Learning (2006 onwards)

Advances such as better activation functions (ReLU), large labeled datasets, GPUs, and improved optimization techniques enabled effective training of deep networks.

Landmark successes include ImageNet (2012), deep speech recognition, and breakthroughs in NLP (e.g., transformers).

Difference Between Deep Learning and Traditional Machine Learning -

| Aspect                          | Traditional Machine Learning               | Deep Learning                               |
| ------------------------------- | ------------------------------------------ | ------------------------------------------- |
| **Feature Engineering**         | Manual feature extraction required         | Features learned automatically              |
| **Model Complexity**            | Shallow models (e.g., SVM, Decision Trees) | Deep multi-layer neural networks            |
| **Data Requirement**            | Works well with small to medium datasets   | Requires very large datasets                |
| **Computational Power**         | Lower computational needs                  | High computational requirements (GPUs/TPUs) |
| **Interpretability**            | Easier to interpret                        | Often considered a “black box”              |
| **Performance on Complex Data** | Limited on images, audio, text             | Excellent performance on unstructured data  |

Question 2: Explain the basic architecture and functioning of a Perceptron. What are its
limitations?

Ans - A Perceptron is the simplest form of an artificial neural network and serves as the foundation of modern neural networks. It is a binary linear classifier used to classify input data into one of two classes. The perceptron was introduced by Frank Rosenblatt in 1958 and is inspired by the functioning of a biological neuron.

2. Basic Architecture of a Perceptron

The perceptron consists of the following components:

Input Layer

Inputs:

x1​,x2​,…,xn

Each input represents a feature of the data.

Weights

Each input has an associated weight w1​,w2​,…,wn

Weights indicate the importance of each input.

Bias

A bias term b is added to shift the decision boundary.

Summation Unit

Computes the weighted sum:

z=i=1∑n​wi​xi​+b

Activation Function

Typically a step (threshold) function:

y = 1 , if z ≥ 0
  0   , if z < 0
	​


Produces the final binary output.

3. Functioning of a Perceptron

Input Processing

Inputs are multiplied by corresponding weights and summed with the bias.

Activation

The weighted sum is passed through the activation function.

Output Generation

The perceptron outputs either 0 or 1, representing class labels.

Learning Process (Training)

Weights are updated using the Perceptron Learning Rule:

winew​=wiold​+η(ytrue​−ypred​)xi​
	​
where
η is the learning rate.

Updates continue until classification errors are minimized.

4. Limitations of the Perceptron

Linear Separability

Can only classify linearly separable data.

Fails on problems like XOR, which are non-linearly separable.

Single-Layer Model

Cannot model complex relationships due to absence of hidden layers.

Binary Output Only

Produces only two classes; not suitable for multi-class problems without modification.

Non-Differentiable Activation

Step function is not differentiable, limiting use of gradient-based optimization.

Sensitivity to Noise

Performs poorly when data is noisy or overlapping.

Question 3: Describe the purpose of activation function in neural networks. Compare

Ans - Purpose of Activation Function in Neural Networks

An activation function determines whether a neuron should be activated or not by introducing non-linearity into the neural network. Without activation functions, a neural network would behave like a simple linear model regardless of the number of layers, and thus would be unable to learn complex patterns.

Key Purposes:

Introduce Non-Linearity – Enables the network to learn complex and non-linear relationships.

Control Output Range – Scales neuron outputs to a specific range.

Enable Deep Networks – Makes multi-layer neural networks expressive and powerful.

Support Learning – Allows gradient-based optimization through backpropagation.

Comparison of Sigmoid, Tanh, and ReLU Activation Functions
1. Sigmoid Activation Function

Definition:

σ(x)=1/1+e^(-x)

Characteristics:

Output range: (0, 1)

Smooth and differentiable

Advantages:

Useful for binary classification output layers

Probabilistic interpretation

Disadvantages:

Vanishing gradient problem

Outputs not zero-centered

Computationally expensive

2. Tanh (Hyperbolic Tangent) Activation Function

Definition:

tanh(x)=ex−e−x​ /ex+e−x

Characteristics:

Output range: (-1, 1)

Zero-centered

Advantages:

Stronger gradients than sigmoid

Faster convergence than sigmoid

Disadvantages:

Still suffers from vanishing gradients

Computationally expensive

3. ReLU (Rectified Linear Unit) Activation Function

Definition:

ReLU(x)=max(0,x)

Characteristics:

Output range: [0, ∞)

Simple and efficient

Advantages:

Reduces vanishing gradient problem

Faster training and convergence

Widely used in deep networks

Disadvantages:

Dying ReLU problem (neurons stuck at zero)

Not zero-centered

Question 4: What is the difference between Loss function and Cost function in neural
networks? Provide examples.

Ans - In neural networks, loss functions and cost functions are used to measure how well a model’s predictions match the actual target values. They guide the learning process by providing feedback that is minimized during training. Although the terms are sometimes used interchangeably, they have distinct meanings in theory.

2. Loss Function
Definition

A loss function measures the error for a single training example by comparing the predicted output with the true label.

Purpose

Evaluates how wrong the model is for one data point

Provides a signal for updating weights during backpropagation

Examples of Loss Functions

Mean Squared Error (MSE) Loss (Single Sample)

L(y,y^​)=(y−y^​)2

Binary Cross-Entropy Loss

L(y,y^​)=−[ylog(y^​)+(1−y)log(1−y^​)]

Hinge Loss (SVM)

L(y,y^​)=max(0,1−yy^​)

Cost Function
Definition

A cost function is the average or total loss over the entire training dataset (or a batch of samples).

Purpose

Represents the overall performance of the model

Used by optimization algorithms (e.g., gradient descent) to update parameters

Examples of Cost Functions

Mean Squared Error (MSE) Cost

(θ)=N1​i=1∑N​(yi​−y^​i​)2

Cross-Entropy Cost Function

J(θ)=−N1​i=1∑N​[yi​log(y^​i​)+(1−yi​)log(1−y^​i​)]

Regularized Cost Function

J(θ)=Loss+λ∑w2

Key Differences Between Loss and Cost Functions -

| Aspect     | Loss Function                        | Cost Function                |
| ---------- | ------------------------------------ | ---------------------------- |
| Scope      | Single training example              | Entire dataset or batch      |
| Purpose    | Measures individual prediction error | Measures overall model error |
| Usage      | Backpropagation per sample           | Optimization objective       |
| Notation   | (L(y, \hat{y}))                      | (J(\theta))                  |
| Dependency | Depends on one data point            | Depends on all data points   |

Relationship Between Loss and Cost Function

The cost function is derived from the loss function.

Cost function is typically the mean or sum of individual losses.

Minimizing the cost function indirectly minimizes individual losses.

Question 5: What is the role of optimizers in neural networks? Compare Gradient

Ans - Role of Optimizers in Neural Networks
1. Introduction

In neural networks, an optimizer is an algorithm used to update model parameters (weights and biases) in order to minimize the cost (objective) function. Optimizers determine how the network learns by controlling the direction and size of weight updates during training.

2. Role of Optimizers

Optimizers play a crucial role in training neural networks by:

Minimizing the Loss/Cost Function

They reduce the error between predicted and actual outputs.

Updating Weights Efficiently

Use gradients computed via backpropagation to adjust parameters.

Controlling Learning Speed

Manage step size through learning rates and adaptive mechanisms.

Ensuring Convergence

Help the model converge faster and avoid local minima or saddle points.

Handling Noisy Gradients

Improve stability when training with mini-batches.

3. Gradient Descent
Description

Gradient Descent (GD) updates parameters in the direction of the negative gradient of the cost function.

Update Rule

w=w−η∇J(w)

Types

Batch Gradient Descent

Stochastic Gradient Descent (SGD)

Mini-batch Gradient Descent

Advantages

Simple and easy to implement

Works well for convex problems

Disadvantages

Slow convergence

Sensitive to learning rate choice

Can get stuck in local minima or saddle points

4. RMSprop (Root Mean Square Propagation)
Description

RMSprop is an adaptive learning rate optimizer that adjusts the learning rate for each parameter individually.

Update Rule

vt​=βvt−1​+(1−β)(∇J(w))2

5. Adam (Adaptive Moment Estimation)
Description

Adam combines the benefits of Momentum and RMSprop by maintaining first and second moment estimates of gradients.

Disadvantages

Higher memory usage

Sometimes leads to poorer generalization compared to SGD.

Question 6: Write a Python program to implement a single-layer perceptron from
scratch using NumPy to solve the logical AND gate.

Ans -



```python
import numpy as np

# AND gate dataset
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([0, 0, 0, 1])

# Initialize weights and bias
weights = np.zeros(X.shape[1])
bias = 0.0
learning_rate = 0.1
epochs = 10

# Step activation function
def step_function(z):
    return 1 if z >= 0 else 0

# Training the perceptron
for epoch in range(epochs):
    for i in range(len(X)):
        z = np.dot(X[i], weights) + bias
        y_pred = step_function(z)
        error = y[i] - y_pred

        # Update weights and bias
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

# Testing
print("Final Weights:", weights)
print("Final Bias:", bias)
print("\nAND Gate Predictions:")
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    print(f"Input: {X[i]} Output: {step_function(z)}")
```

Question 8: Use Keras to build and train a simple multilayer neural network on the
MNIST digits dataset. Print the training accuracy.

Ans -

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize the data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Build the multilayer neural network
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    x_train, y_train,
    epochs=5,
    batch_size=128,
    verbose=1
)

# Print training accuracy
print("Training Accuracy:", history.history['accuracy'][-1])
```

Question 9: Visualize the loss and accuracy curves for a neural network model trained
on the Fashion MNIST dataset. Interpret the training behavior.

Ans -

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist

# Load Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Normalize the data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Build the model
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.2,
    verbose=1
)

# Plot Loss Curve
plt.figure()
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()
plt.show()

# Plot Accuracy Curve
plt.figure()
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy Curve')
plt.legend()
plt.show()
```

Question 10: You are working on a project for a bank that wants to automatically detect
fraudulent transactions. The dataset is large, imbalanced, and contains structured

Explain your real-time data science workflow:

● Which activation function and loss function would you use, and why?

Ans - Problem Context

Data: Large, structured, highly imbalanced

Features: Transaction amount, merchant ID, location, etc.

Task: Binary classification → Fraud (1) vs Legitimate (0)

Constraint: Real-time predictions with high recall for fraud cases

1. Model Design (Perceptron vs Multilayer Neural Network)
Choice: Multilayer Neural Network (MLP)

Why not a single perceptron?

Fraud patterns are non-linear

Structured features interact in complex ways

A single perceptron can only model linear decision boundaries

Model Architecture

Input layer: Numerical & encoded categorical features

Hidden layers: 2–3 Dense layers

Output layer: 1 neuron (binary classification)

Example: Input → Dense(64) → Dense(32) → Output(1)

2. Activation Function & Loss Function

Activation Functions

| Layer         | Activation  | Reason                           |
| ------------- | ----------- | -------------------------------- |
| Hidden layers | **ReLU**    | Fast, avoids vanishing gradients |
| Output layer  | **Sigmoid** | Outputs probability (0–1)        |

Loss Function: Binary Cross-Entropy

L=−[ylog(y^​)+(1−y)log(1−y^​)]

Designed for binary classification.

Works well with sigmoid output.

Interpretable as probability loss.

Training & Evaluation (Handling Class Imbalance)
Challenges

Fraud cases may be < 1%

Accuracy alone is misleading

Solutions

Class Weights

Penalize fraud misclassification more heavily

Evaluation Metrics

Precision

Recall (VERY important for fraud)

F1-score

ROC-AUC

4. Optimizer & Overfitting Prevention

Optimizer: Adam

Adaptive learning rate

Fast convergence

Works well with large datasets and noisy gradients

Preventing Overfitting

Dropout layers

Early stopping

Regularization (L2)

Validation split

Python Code (Keras Implementation)


```python
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# Simulated dataset
X = np.random.rand(10000, 10)   # 10 structured features
y = np.zeros(10000)
y[:500] = 1                     # 5% fraud (imbalanced)
np.random.shuffle(y)

# Compute class weights
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y),
    y=y
)
class_weight_dict = {0: class_weights[0], 1: class_weights[1]}

# Build the model
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(10,)),
    layers.Dropout(0.3),
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(1, activation='sigmoid')
])

# Compile
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy', tf.keras.metrics.Recall(), tf.keras.metrics.Precision()]
)

# Train
history = model.fit(
    X, y,
    epochs=10,
    batch_size=128,
    validation_split=0.2,
    class_weight=class_weight_dict,
    verbose=1
)

# Print final metrics
print("Final Training Accuracy:", history.history['accuracy'][-1])
print("Final Training Recall:", history.history['recall'][-1])
```

“The low accuracy and recall occur because the dataset contains randomly generated features with no correlation to the target labels. Since no meaningful patterns exist, the neural network performs close to random guessing (≈50%). In real fraud detection systems, structured and informative features enable the model to achieve high recall and accuracy.”

A multilayer neural network with ReLU activations, sigmoid output, binary cross-entropy loss, Adam optimizer, and class-imbalance handling is well-suited for real-time fraud detection. Proper evaluation metrics and regularization ensure robust and scalable deployment.
